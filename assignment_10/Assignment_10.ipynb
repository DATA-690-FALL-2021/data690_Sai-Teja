# Tidy Data
Since we have seen the basic methods and functions of the pandas library, let us have a closer look at the main focus of this notebook, i.e. Tidy data

What does Tidy Data Mean? It is a standard way of mapping the meaning of a dataset to its structure. In Tidy Data:

- Each Variable forms a column
- Each observation forms a row.
- Each type of observational unit forms a table.

```python
billboard = pd.read_csv('billboard.csv')

billboard.head()
```

```python
billboard_melt = pd.melt(
    billboard, 
    id_vars = ['year', 'artist', 'track' ,'time','date.entered'],
    var_name= 'week',
    value_name = 'rating' )     #trimming down the multiple week columns to make only one week column using melt function
```

```python
weather = pd.read_csv('weather.csv')
```

```python
weather.head(10)
```

```python
weather_melt = pd.melt(
    weather,
    id_vars=['id','year','month','element'], 
    var_name='day', 
    value_name= 'temp'
    )                   #in this dataframe, we reduced the multiple unnecessary coloumns and just included a single column named day 

weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)               #In this code, we are basically including 2 separate columns for representing max and min temperatures of each day instead of irregular data that we had earlier

weather_tidy.reset_index()
```
