```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/DATA-690-FALL-2021/data690_Sai-Teja/main/Individual_project_part-B/wdi_data%20(5).csv")
df
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
df_2015 = df.query("Year == 2015")
df_2015.sample(10)
```

```python
df_2015 = df_2015.query("indicator in['NE.EXP.GNFS.ZS','NE.TRD.GNFS.ZS','NV.IND.TOTL.ZS','NY.GDP.PCAP.PP.CD']")
df_2015.head()
```

```python
df_pivot = df_2015.pivot(index =['Country Code','Country Name','Region','Income Group','Lending Type'], columns='indicator', values='value').reset_index()
df_pivot.head()
```

```python
df_pivot.info()
```

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="NE.EXP.GNFS.ZS",
    title="TOP 5 Countries by 2015 Exports of goods and services (% of GDP)",
    color='Country Name',
    labels={"NE.EXP.GNFS.ZS":"2015 Exports of goods and services (% of GDP)"}
).update_xaxes(categoryorder="total descending")
fig.show()
```

```python
df2 = df.query("indicator in['NE.EXP.GNFS.ZS','NE.TRD.GNFS.ZS','NV.IND.TOTL.ZS','NY.GDP.PCAP.PP.CD']")
df2.head()
```

```python
df_pivot2 = df2.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns='indicator', values='value').reset_index()
df_pivot2.head()
```

```python
fig = px.line(df_pivot2,
    x="Year",
    y="NV.IND.TOTL.ZS",
    title="TOP 5 Countries by 2015 Industry (including construction), value added (% of GDP)",
    color='Country Name',
    labels={"Year":"Year","NV.IND.TOTL.ZS":"Industry (including construction), value added<br>(% of GDP)"},
    hover_name='Country Name',
    
)
fig.show()
```

```python
fig = px.bar(
    df_pivot,
    x="Country Name",
    y="NE.TRD.GNFS.ZS",
    title="TOP 5 Countries by 2015 Trade (% of GDP)",
    color='Country Name',
    labels={"NE.TRD.GNFS.ZS":"2015 Trade (% of GDP)"}
).update_xaxes(categoryorder="total descending")
fig.show()
```

```python
fig = px.pie(
    df_pivot,
    values='NY.GDP.PCAP.PP.CD', 
    names='Country Name',
    title='TOP 5 Countries by 2015 GDP per capita, PPP (current international $)',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
