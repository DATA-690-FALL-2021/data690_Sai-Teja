```python
def cleanse_pop(a):
  return int(a.replace(",","").replace("*","")) 
df["year"] = df["Census Year"].apply(cleanse_pop)  
df["population"] = df["Total Population"].apply(cleanse_pop)
df
```

```python
def cleanse_pop(a):
  return (a.replace(",","").replace("$","").replace(" Billion","000000000").replace("cents",""))
df["cost"] = df["Census Cost"].apply(cleanse_pop)
df["average"] = df["Average Cost Per Person"].apply(cleanse_pop)
df
```

```python
df['cost'] = pd.to_numeric(df['cost'])
df['average'] = pd.to_numeric(df['average'])
```
